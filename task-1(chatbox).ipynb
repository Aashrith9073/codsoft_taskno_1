```python
from datetime import datetime
import random
```

```python
def chatbot():
    print("Assistant: 🤖 Hi, I am Appu, Nice to meet you!!")
    print("           I am here to help you with some baisc needs.")
    print("           To exit the chatbot you can type 'quit'")
    user = ""
    

    while True:
        prompt = input("you: ")

        if prompt.lower() == "hello" or prompt.lower()=="good morning" or prompt.lower()=="hi":
            print(f"You: {prompt}")
            print("Assistant: Hello! Can i know your good name please? 😊")
            name = input("can i know your good name please?")
            user = name
            print(f"           Hello!! {user} 🤩 It's Nice to meet you!!")
            print("           How can i help you??")
        elif prompt.lower() == "what is your name" or prompt.lower() == "how can i call you":
            print(f"{user}: {prompt}")
            print("Assistant: 😊 My name is Appu, nice to meet you!!")
        elif prompt.lower() == "how are you":
            print(f"{user}: {prompt}")
            print("Assistant: I am a bot, so i don't have feelings ,but i am here to help you in your basic needs 😊")
        elif prompt.lower() == "what can you do":
            print(f"{user}: {prompt}")
            print("Assistant: I can tell you the:")
            print("                             1.Current time⏰")
            print("                             2.A joke😊")
            print("                             3.Calculate few answers🧮")
            print("                             4.Play a game🎮")
        elif prompt.lower() == "tell me a joke":
            print(f"{user}: {prompt}")
            print("Assistant: 😄 Why is the maths book always sad?")
            print("           Because.....It has many problems 🤣")
        elif prompt.lower() == "what is the time":

            print(f"{user}: {prompt}")
            print("Assistant: The ⏰ now is " + datetime.now().strftime("%H:%M:%S"))    
        elif prompt.lower().startswith("calculate"):
            print(f"{user}: {prompt}")
            To_calculated = prompt[9: ]
            result = eval(To_calculated)
            print(f"Assistant: The answer is {result}")
        elif prompt.lower() == "can you play a game":
            print(f"{user}: {prompt}")
            print("Assistant: Yes!! I do have a game for you")   
            print("           It's rock-paper-scissors🤘")    
            print('''           Do you want to play it with me? Type "yes" or "no" ''')
            choice = input("yes or no")
            print(f"           So I take it as {choice}")
            if choice == "yes":
                print("           Make your choice!!")
                user_choice = input("your choice: ")
                print(f"           Your choice is {user_choice}")
                options = ["rock" , "paper" , "scissors"]
                Assistant_choice = random.choice(options)
                print(f"           🤖 Bot choice is {Assistant_choice}")
                if user_choice == Assistant_choice:
                    print("           It's a tie!! 🤝")
                elif (user_choice == "rock" and Assistant_choice == "scissors") or (user_choice == "scissors" and Assistant_choice == "paper") or (user_choice == "paper" and Assistant_choice == "rock"):
                    print("           You win! 🎉")    
                else:
                    print("           I win 😊")    
            else:
                print("Okay, may be next time!!😊")
                      

        elif prompt.lower() == "quit":
            print(f"{user}: {prompt}")
            print("Assistant: It was nice chatting with you! Goodbye!! 👋")
            print("           Hope I will see you hear again!! 😊")
            break
        else:
            print("Assistant: sorry, my data is limited try asking another question")

if __name__ == "__main__":
    chatbot()                                 

```
